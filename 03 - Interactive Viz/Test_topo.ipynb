```python
import folium
import json
import pandas as pd
import random
```

```python
cantons_geo_new = 'data/ch-all-new.geo.json'
cantons_geo = 'data/ch-cantons.topojson.json'

swiss_map = folium.Map(location=[46.801111,8.226667], zoom_start=2)
swiss_map.choropleth(geo_path=cantons_geo, 
                     topojson='objects.cantons', 
                     #line_opacity=0.1, 
                     #fill_opacity=0.3,
                     #fill_color='YlGnBu'
                    )
swiss_map
```

```python
with open(cantons_geo, "r") as fin:
    cantons = json.load(fin)
```

```python
idx = 11
print(cantons['objects']['cantons']['geometries'][idx]['id'])
print(cantons['objects']['cantons']['geometries'][idx]['properties']['name'])
```

```python
list_ids = []
val = []
for i in range(len(cantons['objects']['cantons']['geometries'])):
    list_ids.append(cantons['objects']['cantons']['geometries'][i]['id'])
    val.append(random.random())
    
frame = {'Cantons': list_ids, 'Values':val}
```

```python
df = pd.DataFrame.from_dict(frame)
df.head()
```

```python
swiss_map = folium.Map(location=[46.801111,8.226667], zoom_start=7.7)
swiss_map.choropleth(geo_path=cantons_geo, 
                     data=df,
                     columns=['Cantons', 'Values'],
                     key_on='feature.id',
                     topojson='objects.cantons',
                     line_opacity=0.1, 
                     fill_opacity=0.3,
                     fill_color='YlGn',
                     legend_name = 'Random numbers'
                    )
swiss_map
#swiss_map.save('swiss.html')
```
